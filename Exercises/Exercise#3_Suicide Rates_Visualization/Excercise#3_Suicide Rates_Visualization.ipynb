<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea"> Bivariate Analysis

<font color=darkblue>
    
Through bivariate analysis we try to analyze two variables simultaneously. As opposed to univariate analysis where we check the characteristics of a single variable, in bivariate analysis we try to determine if there is any relationship between two 
variables.

&nbsp;

There are essentially 3 major scenarios that we will come across when we perform bivariate analysis
1. Both variables of interest are qualitative
2. One variable is qualitative and the other is quantitative
3. Both variables are quantitative

&nbsp;

For the purpose of this exercise, we will explore few most popular techniques to perform bivariate analysis.

The following plots are not limited to the headings they are under. They are the options we have if we face a certain scenario.




### Numerical  vs. Numerical
    1. Scatterplot
    2. Line plot
    3. Heatmap for correlation
    4. Joint plot

### Categorical vs. Numerical
    1. Bar chart
    2. Violin plot
    3. Categorical box plot
    4. Swarm plot
    

## Two Categorical Variables 
    1. Bar chart
    2. Grouped bar chart
    3. Point plot

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea"> 1.1 Loading the libraries

```python
# import needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# to suppress warnings
import warnings
warnings.filterwarnings('ignore')
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea"> 1.2 Import the dataset 

```python
# read csv file
data = pd.read_csv("master(1).csv")
```

 <span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea">1.4 Bar plot to check Number of Suicides by top 10 Countries

# This is an example of Numerical vs Categorical.

```python
# Visualizing the top 10 countries with highest total of suicides
data_1 = data.groupby(['country'])['suicides_no'].sum().sort_values(ascending=False).head(10)
data_1.plot.bar(figsize=(15,8),color='pink')
plt.title('Bar Plot (Number of Suicides by top 10 Countries)')
plt.xlabel('Country')
plt.ylabel('Count')
```

## What are the highest three countries in suicides no?

#### Answer:

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea">1.5 Bar plot to check Number of Suicides by 10 bottom Countries

```python
# Visualizing the bottom 10 countries with lowest suicides
data_2 = data.groupby(['country'])['suicides_no'].sum().sort_values(ascending=False).tail(10)
data_2.plot.bar(figsize=(15,8),color='orange')
plt.title('Bar Plot (Number of Suicides by bottom 10 Countries)')
plt.xlabel('Country')
plt.ylabel('Count')
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea">1.6 Bar plot for Number of Suicides Vs Age

```python
# Visualizing a simple view for number of suicides per age.
data_3 = data.groupby(['age'])['suicides_no'].sum().sort_values(ascending=False)
data_3.plot(kind='bar',legend=True,figsize=(8,6),color='blue')
plt.title('Bar Plot (Number of Suicides Vs age)')
plt.xlabel('Age')
plt.ylabel('Count')
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea"> 1.7 Bar plot  Number of Suicides Vs Sex

```python
# Visualizing a simple view of total number of suicides per sex category in all the years
data_4 = data.groupby(['sex'])['suicides_no'].sum().sort_values(ascending=False)
data_4.plot(kind='bar',legend=True,figsize=(8,6),color='green')
plt.title('Bar Plot (Number of Suicides Vs sex) ')
plt.xlabel('Sex')
plt.ylabel('Count')
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea">1.8 Bar plot Number of Suicides Vs generation

```python
# Visualizing a simple view for number of suicides per generation.
data_5 = data.groupby(['generation'])['suicides_no'].sum().sort_values(ascending=False)
data_5.plot(kind='bar',legend=True,figsize=(8,6),color='red')
plt.title('Bar Plot (Number of Suicides Vs generation)')
plt.xlabel('Generation')
plt.ylabel('Count')
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea">1.9 Scatter plot Number of Suicides Vs population

# This is an examples of  Numerical  vs. Numerical

```python
# Visualizing a scatter plot for number of suicides vs population
# 1- Using matplotlib

X = data['suicides_no']
Y = data['population']
plt.title('Scatter Plot (Number of Suicides Vs population)')
plt.xlabel('suicides_no')
plt.ylabel('population')
plt.scatter(X,Y)
plt.show()

# 2- Using seaborn
sns.scatterplot(x='suicides_no',y='population', data=data, size = "population")
plt.title('Scatter Plot (Number of Suicides Vs population)')
plt.xlabel('suicides_no')
plt.ylabel('population')
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea">1.10 Line plot Number of Suicides Vs population only for head() using seaborn

```python
# Visualizing a line plot for number of suicides vs population only for head()
# Using seaborn
sns.lineplot(x='suicides_no',y='population', data=data.head())
plt.title('Line Plot (Number of Suicides Vs population)')
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea">1.11 Scatter plot Number of Suicides/100k Population Vs GDP Per Capita

```python
# Visualizing a scatter plot for number of suicides vs population
# 1- Using matplotlib

X = data['suicides/100k pop']
Y = data['gdp_per_capita ($)']
plt.title('Scatter Plot (Number of Suicides/100k Population Vs GDP Per Capita ($))')
plt.xlabel('suicides/100k pop')
plt.ylabel('gdp_per_capita ($)')
plt.scatter(X,Y)
plt.show()

# 2- Using seaborn
sns.scatterplot(x='suicides/100k pop',y='gdp_per_capita ($)', data=data, size = "gdp_per_capita ($)")
plt.title('Scatter Plot (Number of Suicides/100k Population Vs GDP Per Capita ($))')
plt.xlabel('suicides/100k pop')
plt.ylabel('gdp_per_capita ($)')
```

# Checking trends with Temporal Data 

Temporal data is simply data that represents a state in time, such as the land-use patterns of Hong Kong in 1990,
or total rainfall in Honolulu on July 1, 2009. Temporal data is collected to analyze weather patterns and other environmental variables, monitor traffic conditions, study demographic trends, and so on. This data comes from many sources ranging
from manual data entry to data collected using observational sensors or generated from simulation models.
